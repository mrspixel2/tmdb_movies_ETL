```python
import pandas as pd
import requests
```

# Settings

```python
f = open('Authentification.txt', 'r')
api_key = f.read().strip()
f.close()
api_key
```

# Extract

```python
def getDataFromTMDB(id):
        url = "https://api.themoviedb.org/3/movie/"+ str(id)
        params = {'api_key': api_key}
        response = requests.get(url, params=params)
        return response.json()
```

```python
data = {}
for i in range(0, 6):
    data[i] = getDataFromTMDB(550+i)
data
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
df
```

```python
list(df.columns)
```

# Transform

```python
columns_to_keep = ['budget', 'genres', 'id', 'imdb_id', 'original_title', 'release_date', 'revenue', 'runtime']
df = df[columns_to_keep]
df
```

## Transformation 1

```python
df_copy = df.copy()
genres = pd.DataFrame(df_copy.pop("genres").explode().tolist())
genres = genres.drop_duplicates(subset=['id'])
genres
```

```python
hot = pd.get_dummies(df.pop('genres').explode().str['name']).groupby(level=0).max()
hot
```

```python
df = pd.concat([df, hot], axis=1)
df
```

## transformation 2

```python
df['release_date'] =  pd.to_datetime(df['release_date'], infer_datetime_format=True)
df
```

```python
time = pd.DataFrame({"id": df["id"]})
time['release_date'] = pd.to_datetime(df['release_date'], infer_datetime_format=True)
time = time.join(pd.DataFrame({"day": time["release_date"].dt.day, "month": time["release_date"].dt.month, "year": time["release_date"].dt.year, "day_of_week": time["release_date"].dt.day_name()}))
time
```

### converting runtime to ‘hours’ and ‘minutes

```python
df['runtime'] = pd.to_datetime(df['runtime'], unit='m').dt.strftime('%H:%M')
df
```

### to csv

```python
df.to_csv('movies.csv', index=False)
time.to_csv('datetimes.csv', index=False)
genres.to_csv('genres.csv', index=False)
```

### to phpMyadmin (relational database)

```python
from sqlalchemy import create_engine

import pymysql

sqlEngine = create_engine('mysql+pymysql://root:@localhost', pool_recycle= 3606)
dbConnection = sqlEngine.connect()
```

```python
dbConnection.execute("CREATE DATABASE IF NOT EXISTS tmdb")
dbConnection.execute("ALTER DATABASE tmdb CHARACTER SET = utf8mb4 COLLATE = utf8mb4_general_ci;")
dbConnection.execute("USE tmdb")
```

```python
try:
    movies_table = df.to_sql('movies', dbConnection, if_exists='replace')
    genres_table = genres.to_sql('genres', dbConnection, if_exists='replace')
    datetimes_table = time.to_sql('datetimes', dbConnection, if_exists='replace')
except ValueError as vx:
    print(vx)
except Exception as ex:
    print(ex)
else:
    print("Tables created and populated successfully.")
finally:
    dbConnection.close()
```
